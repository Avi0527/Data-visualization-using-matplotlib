```python
# Line Graph
# ---------------------------------------------------------------------------------------
# A line graph is a simple graph that shows the results in the from of lines.to create a line graph.we need a series of points represented
# by x and y cordinates.A line will connect these points.the function plot() is used to create a line graph as:

# plt.plot(x,y,'colorname')

# Here, the plot() function creates the line graph.x and y represent lists of x and y coordinates to be displayed as points on X and Y axes.
# 'colorname' indicates the color names to be used for the line.
```

```python
#program -> A program to create a line graph to show the profits of a company in various years.

#to display profits of a company year-wise

import matplotlib.pyplot as plt

years = ['2015','2016','2017','2018','2019','2020','2021']
profits = [9,10,10.5,8.9,12.75,7.51,6.45]

#create graph line

plt.plot(years,profits)

#set title and lable
plt.title("ABS Company")
plt.xlabel('years')
plt.ylabel('profits in million RS')

#show the plotline chart
plt.show()
```

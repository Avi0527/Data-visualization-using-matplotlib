```python
#A pie chart shows a circle that is divided into sector and each sector represent a proportion of the whole.

#program :- A program to display a pie chart showing the percentage of expenditures in each department of a company

#to display employee of different depts in pie chart
import matplotlib.pyplot as plt

#take the expenditures of 4 departments

slices = [50,14,24,16]
#take the department names
depts = ['sales','production','hr','finance']

#takes the colors for each department
cols = ['magenta','cyan','brown','gold']

#create a pie chart
plt.pie(slices, labels=depts, colors=cols, startangle=90, explode=(0,0.2,0,0), shadow = True, autopct='%.1f%%')

#set titles and legend
plt.title('WIPRO-DEPARTMENTWISE EXPENDITURE')
plt.legend()

#show the pie chart
plt.show()
```

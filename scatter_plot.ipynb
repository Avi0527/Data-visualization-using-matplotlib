```python
# Scatter Plot
# -----------------------------------------------------------------
# A sactter plot represent a group of points scattered all over X and Y axes.by observing how the points are scattered,we can understand overall
# relationship between them,we can create a scatter plot by using scatter() function of pyplot module as.

# plt.scatter(x,y, color='red')

# Here,x,y represent the x and y coordinates of the points to be sacttered.color='red' indicated that the points are displayed in red color.
```

```python
#program :- A program to create a scatter plot to show the ages and prices of houses in delhi.

#to be display relationship b/w age and price of house at delhi 

import matplotlib.pyplot as plt
import numpy as np

#take age and prices as lists
age = [6,8,9,2,10,2,9,4,11,12,9]
prices = [97,87,82,120,77,122,99,97,74,79,83]

#draw the scatter plot taking age on x and price on Y axis 
plt.scatter(age,prices,color='red')

#disply main title, labels on x and y axis
plt.title("age and prices at delhi")
plt.xlabel("age in year")
plt.ylabel("price in lakhs of rs")

#display the scatter plot
plt.show()
```
